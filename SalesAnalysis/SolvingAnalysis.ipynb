```python
import pandas as pd
import numpy as np
import os
# import matplotlib.pyplot as plt
```

```python
### Merging data into an single files
list_directories = os.listdir('/home/developer/Desktop/Pandas/SalesAnalysis/Sales_Data')
list_directories = [file for file in list_directories if file.endswith('.csv')]
list_of_dataframes = []
for filename in list_directories:
    df = pd.read_csv(f"/home/developer/Desktop/Pandas/SalesAnalysis/Sales_Data/{filename}") 
    list_of_dataframes.append(df)
df = pd.concat(list_of_dataframes,ignore_index=True)
df.columns = df.columns.str.strip()
df
```

```python

df[['Date', 'Time']] = df['Order Date'].str.split(' ',n=1, expand=True)
df
# df['Order Date']
```

```python
# finding Filtering the data
df = df[df['Order ID'].isna() == False]
df.drop('Order Date',axis = 1,inplace= True)
df.reset_index(inplace = True)
```

```python
df
```

```python
df = df[df['Date'].str[0:2]!='Or']
df
```

ADDING MONTH COLUMN

```python
df['month'] = df['Date'].str[0:2]
df['month'] = df['month'].astype("int32")
df
```

Adding a Sales Coloumn

```python
df['Total_Price'] = df['Quantity Ordered'].astype('int32') * df['Price Each'].astype('float32').apply(lambda x: round(x))
df
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Generate example data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create DataFrame
df = pd.DataFrame(data, columns=['Values'])

# Calculate statistics
mean_value = df['Values'].mean()
median_value = df['Values'].median()
std_deviation = df['Values'].std()
skewness = skew(df['Values'])
kurt = kurtosis(df['Values'])

# Print statistics
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_deviation}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

# Plot histogram and density plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Values'], kde=True)
plt.title('Histogram with Density Plot')

plt.subplot(1, 2, 2)
sns.kdeplot(df['Values'], shade=True)
plt.title('Density Plot')

plt.tight_layout()
plt.show()
```
